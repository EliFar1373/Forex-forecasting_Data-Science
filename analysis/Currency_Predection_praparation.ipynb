```python
import yfinance as yf
pair="EURUSD=X"
data=yf.download(pair,start="2015-01-01", end="2025-01-01")
```

```python
# Simplify column names
if isinstance(data.columns, pd.MultiIndex):
    data.columns = [col[0] for col in data.columns]
```

```python
data.head(4)
```

```python
import os

# Set the parent directory to a folder named 'data' one level up
parent_dir = os.path.abspath(os.path.join(os.getcwd(), "..", "data"))

# Make sure the folder exists
os.makedirs(parent_dir, exist_ok=True)

# Full file path
file_path = os.path.join(parent_dir, "forex_datas.csv")

# Save the CSV
data.to_csv(file_path, index=False)

print(f"CSV saved to: {file_path}")
```

```python
data=data.reset_index()
print(data.columns.tolist())
```

```python
from datetime import datetime
import pandas as pd

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
num_cols = ['Close', 'High', 'Low', 'Open', 'Volume']
data[num_cols] = data[num_cols].apply(pd.to_numeric, errors='coerce')
```
