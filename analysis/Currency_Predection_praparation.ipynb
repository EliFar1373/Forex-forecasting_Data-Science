```python
import pandas
```

```python
import yfinance as yf
pair="EURUSD=X"
data=yf.download(pair,start="2015-01-01", end="2025-01-01")
```

```python
data.head(4)
```

```python
data=data.reset_index()
print(data.columns.tolist())
```

```python
data.columns = [f"{p}_{t}" for p, t in data.columns]
```

```python
data = data.rename(columns={'Close_EURUSD=X': 'Close_EURUSD','High_EURUSD=X':'High_EURUSD',
                           'Low_EURUSD=X':'Low_EURUSD','Open_EURUSD=X':'Open_EURUSD',
                           'Volume_EURUSD=X':'Volume_EURUSD','Date_':'Date'})

(data.columns.tolist())
```

```python
from datetime import datetime
import pandas as pd

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
num_cols = ['Close_EURUSD', 'High_EURUSD', 'Low_EURUSD', 'Open_EURUSD', 'Volume_EURUSD']
data[num_cols] = data[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
data.columns.tolist()
```

```python
import sqlalchemy
from sqlalchemy import create_engine

username = "root"
password = input("enter your pass:")
host = "localhost"
port = "3006"
database = "finance"


engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
data.to_sql("CurrencyR",con=engine,if_exists="append",index=False)




```
